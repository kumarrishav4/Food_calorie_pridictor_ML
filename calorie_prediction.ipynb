```python
import pandas as pd
```

```python
data=pd.read_csv('nutrition.csv')
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
data.shape
```

```python
import numpy as np
# Ensuring all values in the 'saturated_fat' column are strings before applying string operations
data['saturated_fat'] = data['saturated_fat'].astype(str)

# Removing 'g' and converting to float
data['saturated_fat'] = (
    data['saturated_fat']
    .str.replace('g', '', regex=False)  # Remove 'g'
    .replace(['', '-', '--'], np.nan) 
    .astype(float)  # Convert to float
)

# Imputing missing values with the median
data['saturated_fat'].fillna(data['saturated_fat'].median(), inplace=True)

# Verifying the changes
data['saturated_fat'].head()
```

```python
data.head()
```

```python
data.columns
```

```python
# Function to clean units and convert to float
def clean_numeric_column(column):
    return (
        column.astype(str)  
        .str.replace(r'[^\d.-]', '', regex=True) 
        .replace('', np.nan)  
        .astype(float)  
    )
# Apply the cleaning function to all columns with numeric-like data in 'data'
for col in data.columns:
    try:
        data[col] = clean_numeric_column(data[col])
    except Exception as e:
        pass
# Confirming that columns in 'data' are now cleaned and numeric
data.info()
```

```python
data.columns
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

if 'name' in data.columns:
    # Using label encoding for simplicity (alternatively, one-hot encoding could be applied)
    from sklearn.preprocessing import LabelEncoder
    label_encoder = LabelEncoder()
    data['name'] = label_encoder.fit_transform(data['name'])

# Correctly selecting multiple columns using a list
X = data[['total_fat', 'cholesterol', 'sodium', 'calcium', 
          'sugars', 'saturated_fat', 'serving_size', 'vitamin_c']]  # Select relevant features
y = data['calories']  # Target column


# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Split the PCA-reduced dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Initialize models
models = {
    "Linear Regression": LinearRegression(),
    "Random Forest Regressor": RandomForestRegressor(random_state=42),
    "Support Vector Regressor (SVR)": SVR(kernel='rbf', C=1.0, epsilon=0.1),
    "K-Nearest Neighbors Regressor (KNN)": KNeighborsRegressor(n_neighbors=5)
}

# Train and evaluate models
results = {}

for model_name, model in models.items():
    # Train the model
    model.fit(X_train, y_train)
    
    # Make predictions
    y_pred = model.predict(X_test)
    
    # Evaluate performance
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    # Store results
    results[model_name] = {"MSE": mse, "MAE": mae, "R2 Score": r2}

# Convert results to a DataFrame for better visualization
import pandas as pd
results_df = pd.DataFrame(results).T
results_df.head()
```

```python
# Calculate training performance metrics for each model
training_results = {}

for model_name, model in models.items():
    # Predictions on training data
    y_train_pred = model.predict(X_train)
    
    # Evaluate training performance
    mse = mean_squared_error(y_train, y_train_pred)
    mae = mean_absolute_error(y_train, y_train_pred)
    r2 = r2_score(y_train, y_train_pred)
    
    # Store results
    training_results[model_name] = {"Train MSE": mse, "Train MAE": mae, "Train R2 Score": r2}

training_results_df = pd.DataFrame(training_results).T

# Combine training and testing results for comparison

training_results_df.head()
```
